```r
library(ggplot2)
x <- sort(runif(1000, 1, 10))
y <- rnorm(1000, 50, 10) * sort(runif(1000, 1, 10))

df <- data.frame(x=x, y=y)

g <- ggplot(
    df,
    aes (
        x = x,
        y = y,
    )
)
g <- g + geom_point(
    shape = 20,
    size = 0.8,
    na.rm = TRUE
)
g <- g + geom_smooth(
    method = "lm"
)
g <- g + xlab("Label X")
g <- g + ylab("Label Y")
g <- g + ggtitle("ggplot2")

plot(g)
```
